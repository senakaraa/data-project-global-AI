```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
df=tips.copy()
print(df.head(10))
print("\n")
```

```python
print(df.describe().T)
```

```python

print(df["sex"].value_counts(),"\n",
      df["smoker"].value_counts(),"\n",
      df["day"].value_counts(),"\n",
      df["time"].value_counts())
```

```python
sns.boxplot(x=df["total_bill"])

#== sns.boxplot(x="total_bill",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
```

```python
#income depending on the number of people
sns.boxplot(x="size",y="total_bill",data=df)
```

```python
#daily total payout by gender
sns.boxplot(x="day",y="total_bill",hue="sex",data=df)
```

```python
sns.catplot(y="total_bill",kind="violin",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",style="time",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="size",size="size",data=df)
```

Linear visualization

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",row="sex",data=df)
```

```python
flights=sns.load_dataset("flights")
df3=flights.copy()
df3.head()
```

```python
df3["year"].value_counts()
```

```python
df3["passengers"].describe()
```

```python
df3=df3.pivot("month","year","passengers");
df3
```

```python
sns.heatmap(df3)
```

```python
sns.heatmap(df3,annot=True,fmt="d")
```

```python
sns.heatmap(data=df3,annot=True,fmt="d",linewidth=.5,cbar=False)
```
